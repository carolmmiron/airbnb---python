```python
#Checking changes on df_listings

df_listings.head()
```

```python
# MERGING DF_LISTINGS AND DF_LISTINGS_ST

# Select specific columns from df_listings
df_listings_subset = df_listings[['listing_id', 'listing_url', 'description', 'host_is_superhost', 
                                  'host_total_listings_count', 'host_verifications', 'amenities', 
                                  'beds', 'accommodates', 'bathrooms_text']]
```

```python
# Merge the two DataFrames on 'listing_id'

df_listings_merged = pd.merge(df_listings_subset, df_listing_st, on='listing_id', how='inner')
```

```python
#Checking data on new df_listings_merged

df_listings_merged.head()
```

```python
# from df_listings_merged calculate the median (50th percentile) of the 'price' column, ignoring NaN values

median_price = df_listings_merged['price'].median()
df_listings_merged['price'].fillna(median_price, inplace=True)
```

```python
# Checking prices correction

missing_values=df_listings_merged.isnull().sum()
print("Missing Values in Each Column:\n", missing_values)
```

```python
# Checking for any missing the df_reviews

missing_values=df_reviews.isnull().sum()
print("Missing Values in Each Column:\n", missing_values)
```

```python
# Checking for duplicates df_reviews

duplicate_rows = df_reviews.duplicated().sum()
print("Number of Duplicate Rows:", duplicate_rows)
```
